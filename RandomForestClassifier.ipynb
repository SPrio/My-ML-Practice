```python
#load libraries w/ iris datset
from sklearn.datasets import load_iris
#load scikit's learn random forest classifier library
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
#loading data from the iris_dataet
iris=load_iris()
#create an obj called iris w/ the feature variable
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()#printing first 5 row
```

```python
#add a new column w/ the species names, this is what we are going to try to predict
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
#view the top 5 rows
df.head()
```

```python
iris.target #factorization depending upon targrt_name we are dividing into 0,1,2 index
```

```python
iris.target_names
```

```python
#divide all rows in the dataset into training set and test set
#we shall generate random numbers w/ in the range of 0 to 1 & those values less than or equal to .75 against them
#we shall put true in a new attribute is_training or false otherwise
df['is_train']=np.random.uniform(0,1,len(df))<=.75
#view the top 5 rows
df.head()
```

```python
#segregation of training & test datasets
#creating two new datframes, one w/ the traing rows and anther one w/ test rows
train, test = df[df["is_train"]==True],df[df["is_train"]==False]
```

```python
print("Number of row in training & test DataFrames are {} and {}"\
      .format(len(train),len(test)))
```

```python
#now preprocessing the data
#create a list of the feature column's names
features=df.columns[:4]
#view features
(features)
```

```python
train
```

```python
#now convert three species category into number of 0,1,2
y=pd.factorize(train['species'])[0]
```

```python
y
```

```python
#now the very importent step. here we shall train the model.
#create a random forest classifier, by convention clf means means a classifier
clf=RandomForestClassifier(n_jobs=2,random_state=0)
#train the classifier to take the training and learn how they relate to the tring y(the species)\
clf.fit(train[features],y)#train[features]-> train attributes and y-> target value
train[features]
```

```python
#thus I have completed the trainig of my learning odel. Now we shall test the performance of our odel testig w/ our test dataset
#apply the classifier we test dataset (which , we never used before)
clf.predict(test[features])
```

```python
#as we are having three ctegories, so the distinct values are 0,1 and 2
#now we shall view probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
```

```python
#consider any row have values e.g: 0.8, 0.1 and 0.1.then it means probability is 80% for category-1 and 10% each for category-3.
```

```python
#evalution of the classifier
#create actual english names for the plants for each predicted plant class
preds=iris.target_names[clf.predict(test[features])]
#view the PREDICTED species for the first five observstions
preds[0:5]
#type(iris.target_names)
```

```python
#view the actual species names for the first five observations
test['species'].head()
```

```python
#creating a confusion matrix
#this confusion matrix will tell us that for how many species were having same predicted species name and number of species where
# they got wrongly predicted
pd.crosstab(test['species'],preds,rownames=['actual species'],\
           colnames=['predicted species'])
```

```python
#view feature importance
#view a list of the features and their importance scores
list(zip(train[features],clf.feature_importances_))
```
