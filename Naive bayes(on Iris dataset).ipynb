```python
#loading data from the iris_dataet
iris=load_iris()
#create an obj called iris w/ the feature variable
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()#printing first 5 row
```

```python
#add a new column w/ the species names, this is what we are going to try to predict
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
#view the top 5 rows
df.head()
```

```python
#the above dataset is used to construct our classifier.
#creating empty dataFrame
flower=pd.DataFrame()
#now we shall create a new person for whom we know
#their feature values but not the gender.
#or goul is to predict the gender
#creating the feature values for this single row
flower['sepal length (cm)']=[5.4]
flower['sepal width (cm)']=[3.2]
flower['petal length (cm)']=[1.0]
flower['petal width (cm)']=[0.1]
#view the data
flower
```

```python
#calculating the prior values
#we shall calculate the prior for both male and feamles genders
#prior can be either constants or probalitydistributions
#In our example is the simply the probability of being the gender
#number of setosa
n_setosa=df['species'][df['species']=='setosa'].count()

#number of versicolor
n_versicolor=df['species'][df['species']=='versicolor'].count()

#number of virginica
n_virginica=df['species'][df['species']=='virginica'].count()


#total rows
total_people = df['species'].count()
print('count of setosa  = {}, versicolor = {} , virginica = {} and all persons = {} '.format(n_setosa,n_versicolor,n_virginica,total_people))
```

```python
#number of male divide by the tota number of candidates
p_setosa=n_setosa/total_people #prior for male

#no. of feamles divided by the toatl number of candidates
p_versicolor=n_versicolor/total_people #prior foe female


#no. of feamles divided by the toatl number of candidates
p_virginica=n_virginica/total_people #prior foe female

print('prior for setosa = {} and versicolor = {} and virginica={}'.format(p_setosa,p_versicolor,p_virginica))
```

```python
#now doing group by gender and calculating the mean of each feature
df_means=df.groupby('species').mean()

#view the values
df_means
```

```python
#now doing group by gender and calculate variance of each feature 
df_varience=df.groupby('species').var()

#view the values
df_varience
```

```python
#mean for setosa candidates
setosa_sepallen_mean  = df_means['sepal length (cm)'][df_means.index=='setosa'].values[0]
setosa_sepalwid_mean  = df_means['sepal width (cm)'][df_means.index=='setosa'].values[0]
setosa_petallen_mean  = df_means['petal length (cm)'][df_means.index=='setosa'].values[0]
setosa_petalwid_mean  = df_means['petal width (cm)'][df_means.index=='setosa'].values[0]

#print('so mean male height= {}, weight = {}, footsize = {}'.format(male_height_mean,male_weight_mean,male_footsize_mean))
```

```python
#mean for versicolor candidates
versicolor_sepallen_mean  = df_means['sepal length (cm)'][df_means.index=='versicolor'].values[0]
versicolor_sepalwid_mean  = df_means['sepal width (cm)'][df_means.index=='versicolor'].values[0]
versicolor_petallen_mean  = df_means['petal length (cm)'][df_means.index=='versicolor'].values[0]
versicolor_petalwid_mean  = df_means['petal width (cm)'][df_means.index=='versicolor'].values[0]

#print('so mean male height= {}, weight = {}, footsize = {}'.format(male_height_mean,male_weight_mean,male_footsize_mean))
```

```python
#mean for virginica candidates
virginica_sepallen_mean  = df_means['sepal length (cm)'][df_means.index=='virginica'].values[0]
virginica_sepalwid_mean  = df_means['sepal width (cm)'][df_means.index=='virginica'].values[0]
virginica_petallen_mean  = df_means['petal length (cm)'][df_means.index=='virginica'].values[0]
virginica_petalwid_mean  = df_means['petal width (cm)'][df_means.index=='virginica'].values[0]

#print('so mean male height= {}, weight = {}, footsize = {}'.format(male_height_mean,male_weight_mean,male_footsize_mean))
```

```python
#varience for male candidates
setosa_sepallen_varience  = df_varience['sepal length (cm)'][df_varience.index=='setosa'].values[0]
setosa_sepalwid_varience  = df_varience['sepal width (cm)'][df_varience.index=='setosa'].values[0]
setosa_petallen_varience  = df_varience['petal length (cm)'][df_varience.index=='setosa'].values[0]
setosa_petalwid_varience  = df_varience['petal width (cm)'][df_varience.index=='setosa'].values[0]
#print('so varience male height= {}, weight = {}, footsize = {}'.format(male_height_varience,male_weight_varience,male_footsize_varience))
```

```python
#varience for male candidates
versicolor_sepallen_varience  = df_varience['sepal length (cm)'][df_varience.index=='versicolor'].values[0]
versicolor_sepalwid_varience  = df_varience['sepal width (cm)'][df_varience.index=='versicolor'].values[0]
versicolor_petallen_varience  = df_varience['petal length (cm)'][df_varience.index=='versicolor'].values[0]
versicolor_petalwid_varience  = df_varience['petal width (cm)'][df_varience.index=='versicolor'].values[0]
#print('so varience male height= {}, weight = {}, footsize = {}'.format(male_height_varience,male_weight_varience,male_footsize_varience))
```

```python
#varience for male candidates
virginica_sepallen_varience  = df_varience['sepal length (cm)'][df_varience.index=='virginica'].values[0]
virginica_sepalwid_varience  = df_varience['sepal width (cm)'][df_varience.index=='virginica'].values[0]
virginica_petallen_varience  = df_varience['petal length (cm)'][df_varience.index=='virginica'].values[0]
virginica_petalwid_varience  = df_varience['petal width (cm)'][df_varience.index=='virginica'].values[0]
#print('so varience male height= {}, weight = {}, footsize = {}'.format(male_height_varience,male_weight_varience,male_footsize_varience))
```

```python
#writing one user defined functions which will calculate p(X/Y)

def p_x_given_y (x, mean_y, var_y):#prob of x given y
    #input ting the arguments into the probability density functin
    p=(1/np.sqrt(2*np.pi*var_y))*np.exp(-(x-mean_y)**2/(2*var_y))
    
    #returning the value of p
    return p
    
```

```python
#now appling bayes classifier to a new data point

#we are calculating numerator part of the posterior of the male if unclassified observation is male
num_posterior_setosa=p_setosa *\
p_x_given_y(flower['sepal length (cm)'][0],setosa_sepallen_mean,setosa_sepallen_varience)\
*p_x_given_y(flower['sepal width (cm)'][0],setosa_sepalwid_mean,setosa_sepalwid_varience)\
*p_x_given_y(flower['petal length (cm)'][0],setosa_petallen_mean,setosa_petallen_varience)\
*p_x_given_y(flower['petal width (cm)'][0],setosa_petalwid_mean,setosa_petalwid_varience)

num_posterior_setosa
```

```python
#now appling bayes classifier to a new data point

#we are calculating numerator part of the posterior of the male if unclassified observation is male
num_posterior_versicolor=p_versicolor *\
p_x_given_y(flower['sepal length (cm)'][0],versicolor_sepallen_mean,versicolor_sepallen_varience)\
*p_x_given_y(flower['sepal width (cm)'][0],versicolor_sepalwid_mean,versicolor_sepalwid_varience)\
*p_x_given_y(flower['petal length (cm)'][0],versicolor_petallen_mean,versicolor_petallen_varience)\
*p_x_given_y(flower['petal width (cm)'][0],versicolor_petalwid_mean,versicolor_petalwid_varience)

num_posterior_versicolor
```

```python
#now appling bayes classifier to a new data point

#we are calculating numerator part of the posterior of the male if unclassified observation is male
num_posterior_virginica=p_virginica *\
p_x_given_y(flower['sepal length (cm)'][0],virginica_sepallen_mean,virginica_sepallen_varience)\
*p_x_given_y(flower['sepal width (cm)'][0],virginica_sepalwid_mean,virginica_sepalwid_varience)\
*p_x_given_y(flower['petal length (cm)'][0],virginica_petallen_mean,virginica_petallen_varience)\
*p_x_given_y(flower['petal width (cm)'][0],virginica_petalwid_mean,virginica_petalwid_varience)

num_posterior_virginica
```

```python
if(num_posterior_setosa>num_posterior_versicolor):
    if(num_posterior_setosa>num_posterior_virginica):
        print("it is a setosa")
    else:
        print("it is a virginica")
else:
    if(num_posterior_versicolor>num_posterior_virginica):
        print("it is a versicolor")
    else:
        print("it is a virginica")
```
