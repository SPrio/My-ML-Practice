```python
#load libraries w/ iris datset
from sklearn.datasets import load_iris
#load scikit's learn random forest classifier library
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
```

```python
#loading data from the iris_dataet
iris=load_iris()
#create an obj called iris w/ the feature variable
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()#printing first 5 row
```

```python
iris.feature_names
```

```python
X=df.copy()
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm
```

```python
iris.target
```

```python
lm.fit(X,iris.target)
```

```python
print('estimated coefficients: ',lm.coef_)
print('estimated intercept: ',lm.intercept_)
```

```python
#lasso regression
#load libraries
from sklearn.linear_model import Lasso
#from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
```

```python
scaler1=StandardScaler()
X_std1=scaler1.fit_transform(X)
scaler1,X_std1
```

```python
#create lasso regression w/ alpha value
regr1=Lasso(alpha=.5)
#fit the linear regression
model1=regr1.fit(X_std1,iris.target)
regr1,model1
```

```python
print('model.coef_: ',model1.coef_)
print('Lasso model.intercept_: ',model1.intercept_)
```
